```python
'''Welcome to the calculator'''

def calculator():
    print("1- Addition")
    print("2- Subtraction")
    print("3- Multiplication")
    print("4- Division")
    print("5- Exit")
    
    while True:
        choice = input('Please choice a function (1/2/3/4/5) : ')
        if choice == '1':   # adds two numbers together as float type
            try:
                def add(a=float(input('Enter a number: ')), b=float(input('Enter a number: '))):
                    print(f'Addition: {a} + {b} =',a+b)
                add()
                again = input('Do you want to calculate again?(y/n): ')
                if again == 'y' or again == 'Y':
                    continue
                else:
                    break
            except Exception as e:
                print("There is a mistake:", e) 

        elif choice == '2':   # subtracts two numbers
            try:    
                def subtract(a=float(input('Enter a number: ')), b=float(input('Enter a number: '))):
                    print(f'Subtraction: {a} - {b} =',a-b)
                subtract()
                again = input('Do you want to calculate again?(y/n): ')
                if again == 'y' or again == 'Y':
                    continue
                else:
                    break
            except Exception as e:
                print("There is a mistake:", e)

        elif choice == '3':   # multiply two numbers
            try:
                def multiply(a=float(input('Enter a number: ')), b=float(input('Enter a number: '))):
                    print(f'Multiplication: {a} x {b} =',a*b)
                multiply()
                again = input('Do you want to calculate again?(y/n): ')
                if again == 'y' or again == 'Y':
                    continue
                else:
                    break
            except Exception as e:
                print("There is a mistake:", e)

        elif choice == '4':   # divides two numbers
            try:
                def divide(a=float(input('Enter a number: ')), b=float(input('Enter a number: '))):
                    print(f'Division: {a} / {b} =',a/b)
                divide()
                again = input('Do you want to calculate again?(y/n): ')
                if again == 'y' or again == 'Y':
                    continue
                else:
                    break
            except Exception as e:
                print("There is a mistake:", e)

        elif choice == "5":  # exits the program
            print("Exiting the program...")
            break

        else:
            print("Invalid input. If you want to exit the program, please enter '5'. ")

calculator()    
```
